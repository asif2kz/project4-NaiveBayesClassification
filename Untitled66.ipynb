## 2. Implement Naive Bayes on your own from scratch for text classification.

```python
## implement own naive bayes.

## fit function will create the dictionary for calculating the probablities.

def fit(x_train,y_train):
    result={}
    
    ## to hold the total feature count for calculating the probablity of a class.
    ## here p(class)=sum(all features in that class) / toatal features sum in all classes.
    result["total_data"] = 0
    
    
    ## take all distinct o/p classes
    opclasses=set(y_train)
    ## go through all o/p classes
    for curr_op_class in opclasses:
        
        
        ## create a dictionary in the following format.
        
        ##  .o/p1 --->feature1 ---->count of each feature
        ##      --->feature2 ---->count of each feature
        ##     --->feature3  ---->count of each feature
        ##     --->total_count.
        
        ##  .o/p2 --->feature1 ---->count of each feature
        ##      --->feature2 ---->count of each feature
        ##     --->feature3  ---->count of each feature
        ##       --->total_count.
        
        ## .total_data    
        
        result[curr_op_class]={}
        curr_class_rows=(y_train==curr_op_class)
        curr_class_x=x_train[curr_class_rows]
        curr_class_y=y_train[curr_class_rows]
        result[curr_op_class]['total_count']=0
        available_features=x_train.shape[1]
        for j in range(available_features):
            result[curr_op_class][j]=curr_class_x[:,j].sum()
            result[curr_op_class]['total_count']+=result[curr_op_class][j]
        result['total_data']+=result[curr_op_class]['total_count']
    return result
```

```python
## function to calculate the probablity of a xpoint in a given class
def probablity(dictionary,x,current_class):
    ## first find probablity of p(current_class)
    ## output= dictionary[current_class][ttotal_count] / dictionary['total_data']
    ## as per application of log the above formula will come as following
    
    
    output= np.log(dictionary[current_class]['total_count']) - np.log(dictionary['total_data'])
    
    
    num_features=len(x)
    for j in range(num_features):
        xj=x[j]
        ## if in given x for feature j if its zero there is no need to consider that.
        if(x[j]==0):
            continue
        ## else first find that features count in curr_class.
        current_class_xj=dictionary[current_class][j]
        ## then find the total feature count in that class.
        current_class_total_count=dictionary[current_class]['total_count']
        ## apply log formuka on that
        current_xj_prob = np.log(current_class_xj) - np.log(current_class_total_count)
        ## add up to final result.
        output += current_xj_prob
    return output
```

```python
def predict_single_point(dictionary,x):
    key=dictionary.keys()
    
    ## initialise the bestclass and its probablity is undifined.
    best_class=-1
    best_probablity=-1000
    
    
    first_run=True
    ## go through each keys in dictionary.
    for k in key:
        
        ## as key contain total_data s key as part of our implementation skip it.
        if k=='total_data':
            continue
            
        ## find the probablity that this point x belong to a particular class.
        p_current_class=probablity(dictionary,x,k)
        
        ## if we get a probablity greater than best_probablity change class to that key and probablity respectivly.
        if(first_run or best_probablity<p_current_class):
            best_class=k
            best_probablity=p_current_class
        first_run=False
    
    
    
    
    return best_class
```

```python
## return the predicted values for test datas.
def predict(dictionary,x_test):
    ypred=[]
    for x in x_test:
        x_class = predict_single_point(dictionary,x)
        ypred.append(x_class)
    return ypred
```

```python
## call fit functio to creayte the dictionary.
dictionary=fit(x_train,y_train)
## then call predict function with genreted dictionary and test data
y_pred_new = predict(dictionary,x_test)
```

```python
## genetate report for userdifind naive nbayes 
print("classification report")
print(confusion_matrix(y_test,y_pred_new))
own_classifier=classification_report(y_test,y_pred_new)
print(own_classifier)
```
